# **Separar Nubes de Puntos**

Uno de los problemas más interesantes en el área de Geometría Computacional es el problema de la *detección de contactos*. Este problema consiste en detectar el punto de contacto o colisión entre dos objetos, ya sean convexos o cóncavos. Existen variados algoritmos que intentan resolver este problema de forma eficiente. Aún así, la mayoría de estos algoritmos considera dos fases muy importantes: 

- **(i)** identificar que existe una posible colisión.
- **(ii)** calcular el punto de contacto propiamente tal. 

Claramente, el paso **(i)** es un tipo de filtro para el paso **(ii)**, ya que este último es muy costoso en términos computacionales.

Una de las técnicas que se propone en la literatura para el paso **(i)** es calcular el **plano común** entre 2 objetos, es decir, calcular un plano que se encuentra entre los objetos para luego verificar si hay un *overlap* de alguno de ellos con el plano. Es decir, si uno de los objetos intersecta el plano común, entonces hay una posible oclisión. 

Cada objeto $k$ se representa como una nube de puntos $\bigg(x_i^{(k)},y_i^{(k)} \bigg)$ para $i \in \{1,...,n_k\}$ donde $n_k$ es la cantidad de puntos que representan al objeto. Dadas **2 nubes de puntos** se busca la **recta común** que las une, es decir, $k \in \{1,2\}$

Ahora surge la pregunta **¿cómo obtenemos la recta común?** si consideramos que cada objeto se representa como una nube de puntos, lo primero que podemos pensar es utilizar **mínimos cuadrados** para obtener una **única** recta $y = \hat{a}x + \hat{b}$ que minimice la distancia para ambas nubes de puntos. Es decir $\bigg(x_i^{(1)},y_i^{(1)} \bigg)$ para $i \in \{1,...,n_1\}$ y $\bigg(x_i^{(2)},y_i^{(2)} \bigg)$ para $i \in \{1,...,n_2\}$

El problema es que no necesariamente la recta se encuentra entre ambos objetos, como se ve a continuacion:

Ya que simplemente se está calculando una regresión lineal con ambas nubes de puntos. *Para subsanar esta dificultad, podemos "forzar" que la recta común esté entre ambos objetos*. Para esto. **aumentamos en una dimensión el problema**, agregando información sobre los objetos (a través del aumento de dimensionalidad).

En este caso el **objeto 1** que está representado por la primera nube sé puntos queda de la siguiente forma: $\bigg(x_i^{(1)},y_i^{(1)}, 1 \bigg)$ y el **objeto 2** con la segunda nube así $\bigg(x_j^{(2)},y_j^{(2)}, -1 \bigg)$, se busca **utilizar mínimos cuadrados para obtener el plano $z = ax + by + c$ que se encuentre más cerca a cada conjunto**, por lo que se busca minimizar el siguiente error:

$$
E_2(a,b,c) = \sum_{i=1}^{n_1} \bigg(1 - (ax_i^{(1)} + b y_i^{(1)} + c) \bigg)^2 + \sum_{j=1}^{n_2} \bigg( -1 - (ax_j^{(2)} + b y_j^{(2)} + c) \bigg)^2$$

En este caso, estamos interesados en el plano proyectado sobre $x − y$, entonces si $z = 0$ se obtiene que la recta común es:

$$0 = ax + by + c \implies by = -ax -c \implies y = \frac{a}{b}x-\frac{c}{b}$$

Lo cual resulta en una estimación mucho mejor que la anterior.

Sin embargo, cuando hay un desequilibrio en la cantidad de puntos $n1$ y $n2$, la recta común no se encuentra necesariamente entre ambos objetos. El desafío de encontrar una recta entre los puntos tiene que ver con el parámetro $\frac{n1}{n2}$ que balancea la búsqueda de mínimos cuadrados, por lo que se aplica la siguiente fórmula:

$$ E_2(a, b, c) = \sum_{i=1}^{n_1} \left(1 - a x^{(1)}_i - b y^{(1)}_i - c \right)^2 + \frac{n_1}{n_2} \sum_{j=1}^{n_2} \left(-1 - a x^{(2)}_j - b y^{(2)}_j - c \right)^2 $$

# **Interpolaciones derivada**

Los algoritmos de interpolación polinomial nos permiten aproximar funciones por medio de una secuencia de evaluaciones de operaciones elementales. Dentro de los algoritmos que requieren menos operaciones se encuentra la **interpolación baricéntrica**, la cual se describe matemáticamente de la siguiente forma:

$$p(x) = \frac{\sum_{i=1}^{n} \left[ y_i \frac{w_i}{(x - x_i)}\right]}{\sum_{i=1}^{n} \left[ \frac{w_i}{(x - x_i)}\right]}$$

donde:
- $x_i$ son los **nodos de interpolación**,
- $y_i = f(x_i)$ es la evaluación de la función en el nodo $x_i$,
- $w_i$ son **pesos** que dependen únicamente de los nodos $x_i$, se define como: $w_i = \frac{1}{\ell_i(x_i)} = \frac{1}{\ell'(x_i)}$
- $n$ es la cantidad total de puntos utilizados para la interpolación.

En general, al interpolar funciones uno puede elegir libremente los puntos donde se interpola la data, pero no el dominio en el cual se quiere interpolar. En esta pregunta, por simplicidad, se considera el dominio $[-1, 1]$, es decir, $x \in [-1, 1]$. Entonces, los candidatos naturales a utilizar como nodos de interpolación son los famosos **puntos de Chebyshev**.

En resumen, el párrafo anterior sugiere que una muy buena estrategia para interpolar polinomialmente una función en el dominio $[-1, 1]$ es usar la **interpolación baricéntrica con puntos de Chebyshev**.

---

### **Ejemplo**

Si se desea interpolar la función $f(x) = \sin(2x)$ en el dominio $[-1, 1]$ con $n$ puntos de Chebyshev $x_i$, se define:

$$
y_i = \sin(2x_i)
$$

Y el polinomio interpolador se construye como:

$$
\hat{p}(x) = \frac{\sum_{i=1}^{n} \left[ \sin(2x_i) \frac{w_i}{(x - x_i)} \right] }{\sum_{i=1}^{n} \left[ \frac{w_i}{(x - x_i)} \right]}
$$

Ahora, si se desea construir una aproximación polinomial de la **derivada** de $f(x)$, basta con derivar:

$$
f'(x) = 2 \cos(2x)
$$

Y luego interpolar esa función, resultando en el polinomio interpolador:

$$
\check{p}(x) = \frac{\sum_{i=1}^{n} \left[ 2 \cos(2x_i) \frac{w_i}{(x - x_i)} \right] }{\sum_{i=1}^{n} \left[ \frac{w_i}{(x - x_i)} \right]}
$$

Por lo tanto:
- $f(x)$ se aproxima polinomialmente con $\hat{p}(x)$,
- $f'(x)$ se aproxima polinomialmente con $\check{p}(x)$.

Ambas aproximaciones usan $n$ puntos y resultan en polinomios de grado a lo sumo $n - 1$.

---

### Caso más general

Este procedimiento es directo y conveniente **si se conoce explícitamente** la función $f(x)$ y su derivada $f'(x)$. Pero en este laboratorio se analiza la situación donde:

1. No se conoce $f(x)$ ni $f'(x)$,
2. Solo se tiene acceso a una **aproximación polinomial** de $f(x)$, donde $y_i = f(x_i)$, y lo único que se conoce son los valores $y_i$, no la función $f$ como expresión. Es decir:

$$p(x) = \frac{\sum_{i=1}^{n} \left[ y_i \frac{w_i}{(x - x_i)}\right]}{\sum_{i=1}^{n} \left[ \frac{w_i}{(x - x_i)}\right]}$$


3. **Se desea aproximar polinomialmente la derivada $f'(x)$, usando $p(x)$.**

---

### Reescritura de la interpolación baricéntrica

Para facilitar el análisis, se reescribe la interpolación baricéntrica así:

$$
p(x) = \frac{\sum_{i=1}^{n} \left[ y_i \frac{w_i}{(x - x_i)}\right]}{\sum_{i=1}^{n} \left[ \frac{w_i}{(x - x_i)}\right]} = \ell(x) \sum_{i=1}^{n} \left[ y_i \frac{w_i}{x - x_i}\right], \quad \text{donde} \quad \ell(x) = \prod_{j=1}^{n} (x - x_j)
$$

Como los nodos $x_i$, los pesos $w_i$, y los valores $y_i$ son conocidos, esta expresión define completamente el polinomio interpolador $p(x)$.

---

### Derivada de la interpolación baricéntrica

Se puede derivar esta expresión para obtener $p'(x)$:

$$
p'(x) = \ell'(x) \left( \sum_{i=1}^{n} \left[ y_i \frac{ w_i}{x - x_i} \right] \right) + \ell(x) \frac{d}{dx} \left( \sum_{i=1}^{n} \left[ y_i \frac{w_i}{x - x_i}\right] \right)
$$

Donde omitiendo algunos pasos por la simplicidad del problema se obtiene:

$$
p'(x) = \sum_{i=1}^{n} \left[ y_i w_i \left( \sum_{j=1, j \ne i}^{n} \left[ \ell_{j,i}(x)\right] \right) \right]
$$

donde:
- $\ell_j(x) = \prod_{k=1, k \ne j}^{n} (x - x_k)$
- $\ell_{j,i}(x) = \prod_{k=1, k \ne j, k \ne i}^{n} (x - x_k)$

---

### Costo computacional

La expresión anterior para $p'(x)$ requiere, en general, **$O(n^3)$** operaciones elementales para evaluar $p'(x)$ en un solo punto $x$. Entonces, si se quiere evaluar $p'(x)$ en $m \gg 1$ puntos, se necesitarán $O(mn^3)$ operaciones.

Por contraste, evaluar $p(x)$ en $m$ puntos solo requiere $O(mn)$ operaciones.

---

### ¿Se puede hacer mejor?

**Pregunta clave:**  
¿Es posible evaluar $p'(x)$ en $m$ puntos utilizando significativamente menos de $O(mn^3)$ operaciones?

La respuesta es que **sí**, usando Chebyshev. Existe una simplificación importante al evaluar $p'(x)$ **en los mismos nodos de Chebyshev** $x_k$:

$$
p'(x_k) = y_k w_k \left( \sum_{j=1, j \ne k}^{n} \left[ \ell_{j,k}(x_k) \right] \right) + \sum_{i=1, i \ne k}^{n} \left[ y_i w_i \ell_{k,i}(x_k) \right]
$$

Esta fórmula tiene costo computacional **$O(n^2)$** por nodo, lo cual es significativamente menor que $O(n^3)$ en el caso general.  
(Notar que se cambió la notación de $x_i$ a $x_k$ solo por claridad, ya que en la sumatoria está el valor i, pero los puntos son los de Chebyshev en igual que en los casos anteriores.)
